```python
def get_yahoo_finance_news(ticker_symbol, limit=10):
    """
    Fetches recent Yahoo Finance news for a given ticker.
    Returns a pandas DataFrame with title, publisher, and URL.
    """
    ticker = yf.Ticker(ticker_symbol)
    news_items = ticker.news
    dex = [i for i in range(1,1)]
    

    if not news_items:
        print(f"No news found for {ticker_symbol}.")
        return pd.DataFrame()

    df = pd.DataFrame(dict(islice(news_items[0]['content'].items(), 6)) , index=dex)
    rows = []
    for i in news_items:
        rows.append(dict(islice(i['content'].items(), 6)))

    df = pd.concat([df, pd.DataFrame(rows)], ignore_index=True)
    
    # Convert publish time from timestamp to readable date
    # df['providerPublishTime'] = pd.to_datetime(df['providerPublishTime'], unit='s')
    
    return df
```

```python
def nvda_stock_news_years(year):
    list_1 = []
    for start_month, end_month in zip(range(1,12),range(2,13)):
        if len(str(start_month)) == 1:
            start_month = f'0{start_month}' 
        if len(str(end_month)) == 1:
            end_month = f'0{end_month}'
        if year < 2019:
            df = pd.read_csv(f'../data/NVDA_{year}_{start_month}_stock_news.csv')
            list_1.append(df)
        else:
            df = pd.read_csv(f'../data/NVDA_{year}_{start_month}_to_{end_month}_stock_news.csv')
            list_1.append(df)
    return list_1
```

```python
#put into function straight

NVDA_2017_data_list = nvda_stock_news_years(2017)
NVDA_2018_data_list = nvda_stock_news_years(2018)
NVDA_2020_data_list = nvda_stock_news_years(2020)
NVDA_2021_data_list = nvda_stock_news_years(2021)
NVDA_2022_data_list = nvda_stock_news_years(2022)
NVDA_2023_data_list = nvda_stock_news_years(2023)
NVDA_2024_data_list = nvda_stock_news_years(2024)
```

```python
import time

def get_news_data(start_year,end_year,company):
    datasets = []
    start_time = time.perf_counter()
    url = "https://api.gdeltproject.org/api/v2/doc/doc"
    
    for year in range(start_year,end_year):
        print(year)
        print("--------------------")
        for start_month, end_month in zip(range(1,12),range(2,13)):
            if len(str(start_month)) == 1:
                start_month = f'0{start_month}' 
            if  len(str(end_month)) == 1:
                end_month = f'0{end_month}' 
                
            params = {
                "query": company,
                "mode": "ArtList",
                "maxrecords": "200",
                "format": "json",
                "startdatetime": f"{year}{start_month}01000000",
                "enddatetime": f"{year}{end_month}31000000"
            }
            
            headers = {"User-Agent": "Mozilla/5.0"}
            time.sleep(2)
            
            response = requests.get(url, params=params, headers=headers)
            # print(response.status_code)
            data = response.json().get('articles', [])
            df_1 = pd.DataFrame(data)[['seendate','title','url','sourcecountry','language']]
            datasets.append(df_1)
            # print(start_month, end_month)
    df = pd.concat(datasets,axis=0,ignore_index=True)
    df['Date'] = pd.to_datetime(df['seendate'], errors='coerce')
    df.drop('seendate', axis=1,inplace=True)
    # df['Date'] = df['Date'].dt.date
    df.sort_values(by='Date',inplace=True)
    end_ttime = time.perf_counter()
    
    elapsed_time = end_ttime - start_time
    print("-------------")
    print(f'code excuted in: {elapsed_time:.4f} seconds')
    return df
```

```python
df_MSFT_2019 = get_news_data(2019,2020,'microsoft')
```

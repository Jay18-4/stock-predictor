```python
import pandas as pd
```

```python
df_stock = pd.read_csv("../data/engineered_interleaved_features_multi_stock_data.csv")
df_stock = df_stock.drop("Unnamed: 0",axis=1)
df_stock.rename(columns={'Date':'date'},inplace=True)
df_stock.head()
```

```python
df_news = pd.read_csv("../data/News_sentiment.csv")
df_news.head()
```

```python
def compute_polarity(row):
    if row['sentiment'] == 'positive':
        return row['score'] * 1.0
    elif row['sentiment'] == 'neutral':
        return row['score'] * 0.5
    else:  # negative
        return row['score'] * 0.0

df_news['sentiment_polarity'] = df_news.apply(compute_polarity, axis=1)
df_news.head()
```

```python
df_dummies = pd.get_dummies(df_news['sentiment'])
df_dummies.head()
```

```python
df_news_sent = pd.concat([df_news,df_dummies],axis=1)
# df_news = df_news.drop("sentiment",axis=1)
df_news_sent.rename(columns={'negative':'sentiment_neg', 'neutral':'sentiment_neu','positive':'sentiment_pos'},inplace=True)
df_news_sent = df_news_sent[['date','sentiment_polarity','sentiment_neg','sentiment_neu','sentiment_pos']]
df_news_sent.head()
```

```python
df_APPL = pd.read_csv("../data/apple_news_data.csv",)
df_APPL.head()
```

```python
df_APPL = df_APPL[['date','sentiment_polarity','sentiment_neg','sentiment_neu','sentiment_pos']]
df_APPL.head()
```

```python
df_news_sentiment = pd.concat([df_APPL,df_news_sent],axis=0)
df_news_sentiment = df_news_sentiment.rename(columns={'date':'datetime'})
df_news_sentiment.head()
```

```python
df_news_sentiment.tail()
```

```python
print(df_stock.shape,df_news_sentiment.shape)
```

```python
df_news_sentiment['datetime'] = pd.to_datetime(df_news_sentiment['datetime'], format="mixed", utc=True)
df_news_sentiment = df_news_sentiment.set_index('datetime').sort_index()
df_news_sentiment.info()
```

```python
df_stock['date'] = pd.to_datetime(df_stock['date'])
df_stock = df_stock.set_index('date').sort_index()
df_stock.info()
```

```python
df_news_sentiment['date'] = df_news_sentiment.index.date
df_news_sentiment['date'] = pd.to_datetime(df_news_sentiment['date'])
df_news_sentiment.head()
```

```python
df_news_sentiment.info()
```

```python
merged_intraday = df_news_sentiment.merge(
    df_stock,
    left_on="date",
    right_index=True,
    how="left"
)
merged_intraday.head()
```

```python
daily_sentiment = (
    df_news_sentiment.groupby('date')
    .agg({
        'sentiment_polarity': ['mean', 'sum'],  # overall sentiment per day
        'sentiment_pos': 'sum',
        'sentiment_neg': 'sum',
        'sentiment_neu': 'sum'
    })
)
daily_sentiment.columns = [
    'pol_mean', 'pol_sum',
    'pos_count', 'neg_count', 'neu_count'
]
daily_sentiment.head()
```

```python
daily_sentiment.tail()
```

```python
daily_sentiment.info()
```

```python
stock_with_sentiment = df_stock.merge(
    daily_sentiment,
    left_index=True,
    right_index=True,
    how="left"
)
stock_with_sentiment.head()
```

```python
stock_with_sentiment['has_news'] = (~stock_with_sentiment['pol_mean'].isna()).astype(int)
sentiment_cols = ['pol_mean','pol_sum','pos_count','neg_count','neu_count'] 
stock_with_sentiment[sentiment_cols] = stock_with_sentiment[sentiment_cols].fillna(0)
stock_with_sentiment.head()
```

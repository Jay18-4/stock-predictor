```python
from deep_translator import GoogleTranslator
import time

translator = GoogleTranslator(source='auto', target='en')

def batch_translate(texts, batch_size=20):
    translator = GoogleTranslator(source='auto', target='en')
    results = []
    for i in range(0, len(texts), batch_size):
        chunk = texts[i:i+batch_size]
        joined = " ||| ".join(chunk)
        try:
            translated = translator.translate(joined)
            results.extend(translated.split(" ||| "))
        except Exception:
            time.sleep(2)
            results.extend([""] * len(chunk))
    return results

# sampled_non_english["translated_title"] = batch_translate(sampled_non_english["title"].tolist())
# sampled_non_english = sampled_non_english.dropna(subset=['translated_title'])
# sampled_non_english['language'] = 'English'
# sampled_non_english['translation_flag'] = True
```

```python
import time
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer, pipeline

start_time = time.perf_counter()

model_path = "C:/huggingface_cache/opus-mt-mul-en"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_path)

def translate(text):
    inputs = tokenizer(text, return_tensors="tf")
    outputs = model.generate(**inputs)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Test it
print(translate("Bonjour tout le monde"))
end_ttime = time.perf_counter()
    
elapsed_time = end_ttime - start_time
print("-------------")
print(f'code excuted in: {elapsed_time:.4f} seconds')
```

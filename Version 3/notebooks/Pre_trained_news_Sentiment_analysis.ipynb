```python
import time

def get_news_data(start_year,end_year,company):
    datasets = []
    start_time = time.perf_counter()
    url = "https://api.gdeltproject.org/api/v2/doc/doc"
    
    for year in range(start_year,end_year):
        print(year)
        print("--------------------")
        for start_month, end_month in zip(range(1,12),range(2,13)):
            if len(str(start_month)) == 1:
                start_month = f'0{start_month}' 
            if  len(str(end_month)) == 1:
                end_month = f'0{end_month}' 
                
            params = {
                "query": company,
                "mode": "ArtList",
                "maxrecords": "200",
                "format": "json",
                "startdatetime": f"{year}{start_month}01000000",
                "enddatetime": f"{year}{end_month}31000000"
            }
            
            headers = {"User-Agent": "Mozilla/5.0"}
            time.sleep(2)
            
            response = requests.get(url, params=params, headers=headers)
            # print(response.status_code)
            data = response.json().get('articles', [])
            df_1 = pd.DataFrame(data)[['seendate','title','url','sourcecountry','language']]
            datasets.append(df_1)
            # print(start_month, end_month)
    df = pd.concat(datasets,axis=0,ignore_index=True)
    df['Date'] = pd.to_datetime(df['seendate'], errors='coerce')
    df.drop('seendate', axis=1,inplace=True)
    # df['Date'] = df['Date'].dt.date
    df.sort_values(by='Date',inplace=True)
    end_ttime = time.perf_counter()
    
    elapsed_time = end_ttime - start_time
    print("-------------")
    print(f'code excuted in: {elapsed_time:.4f} seconds')
    return df
```

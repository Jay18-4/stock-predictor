```python
import pandas as pd
```

```python
df = pd.read_csv("../data/explored_stock_data.csv",index_col="Date")
df.drop("Unnamed: 0",axis="columns",inplace=True)
df.head()
```

```python
df["Net_result"] = df.Close - df.Open
df["Close_Open_Ratio"] = df.Close / df.Open
df['Day_to_Day_Change'] = df['Close'] / df['Close'].shift(1)
df['MA_20'] = df['Close'].rolling(20).mean()
df['Close_MA20_Ratio'] = df['Close'] / df['MA_20']
```

```python
df.drop(["High","Low","Open"],axis="columns",inplace=True)
df.head()
```

```python
df['Rolling_mean_20'] = df['Close'].rolling(window=20, min_periods=1).mean()
df.head()
```

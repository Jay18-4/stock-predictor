```python
import yfinance as yf
import pandas as pd
```

```python
ticker = "AAPL"   # You can change to TSLA, MSFT, NVDA, etc.
data = yf.download(ticker, start="2018-01-01", auto_adjust=True)

print(data.head())
data.to_csv("../data/raw_stock_data.csv")
```

```python
df = pd.read_csv("../data/raw_stock_data.csv", index_col=0, parse_dates=True, date_format="%Y-%m-%d")
df = df.dropna().drop_duplicates()
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df.index, df["Close"])
plt.title(f"{ticker} Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price ($)")
plt.show()
```

```python
df.head()
```

```python
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)
```

```python
df.drop(["Ticker"],axis="rows",inplace=True)
```

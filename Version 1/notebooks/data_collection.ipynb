```python
import yfinance as yf
import pandas as pd
```

```python
ticker = "AAPL"   # You can change to TSLA, MSFT, NVDA, etc.
data = yf.download(ticker, start="2015-01-01", auto_adjust=True)

print(data.head())
data.to_csv("../data/raw_stock_data.csv")
```

```python
df = pd.read_csv("../data/raw_stock_data.csv", index_col=0, parse_dates=True, date_format="%Y-%m-%d")
df = df.dropna().drop_duplicates()
df.describe()
```
